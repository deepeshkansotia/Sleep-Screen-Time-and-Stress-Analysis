```python
df['screen_addiction_level'] = pd.cut(
    df['daily_screen_time_hours'],
    bins=[0,3,6,9,24],
    labels=['Low','Moderate','High','Extreme']
)
```

```python
df['sleep_category'] = pd.cut(
    df['sleep_duration_hours'],
    bins=[0,5,7,9,24],
    labels=['Poor','Average','Healthy','Oversleep']
)
```

```python
df['stress_category'] = pd.cut(
    df['stress_level'],
    bins=[0,3,6,10],
    labels=['Low','Medium','High']
)
```

## VISUALIZATIONS

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['font.size'] = 12
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)
plt.title("Correlation Between Lifestyle Factors, Sleep, and Stress")
plt.show()
```

```python
sns.scatterplot(
    x='daily_screen_time_hours',
    y='stress_level',
    data=df,
    alpha=0.5
)

plt.title("Relationship Between Screen Time and Stress Level")
plt.xlabel("Daily Screen Time (hours)")
plt.ylabel("Stress Level")

plt.show()
```

```python
sns.histplot(
    df['daily_screen_time_hours'],
    bins=30,
    kde=True
)

plt.title("Distribution of Daily Screen Time")
plt.xlabel("Screen Time (hours)")
plt.ylabel("Frequency")

plt.show()
```

```python
occupation_stress = df.groupby('occupation')['stress_level'].mean().sort_values()

sns.barplot(
    x=occupation_stress.values,
    y=occupation_stress.index
)

plt.title("Average Stress Level by Occupation")
plt.xlabel("Average Stress Level")
plt.ylabel("Occupation")

plt.show()
```

```python
sns.scatterplot(
    x='physical_activity_minutes',
    y='stress_level',
    data=df,
    alpha=0.4
)

plt.title("Physical Activity vs Stress Level")
plt.xlabel("Physical Activity (minutes/day)")
plt.ylabel("Stress Level")

plt.show()
```

```python
sns.scatterplot(
    x='sleep_duration_hours',
    y='sleep_quality_score',
    data=df,
    alpha=0.5
)

plt.title("Sleep Duration vs Sleep Quality")
plt.xlabel("Sleep Duration (hours)")
plt.ylabel("Sleep Quality Score")

plt.show()
```

```python
sns.kdeplot(
    df['stress_level'],
    fill=True
)

plt.title("Stress Level Distribution")
plt.xlabel("Stress Level")

plt.show()
```

```python
sns.boxplot(
    x='caffeine_intake_cups',
    y='sleep_duration_hours',
    data=df
)

plt.title("Caffeine Intake vs Sleep Duration")
plt.xlabel("Cups of Caffeine")
plt.ylabel("Sleep Duration (hours)")

plt.show()
```

```python
sns.scatterplot(
    x='phone_usage_before_sleep_minutes',
    y='sleep_quality_score',
    data=df,
    alpha=0.4
)

plt.title("Phone Usage Before Sleep vs Sleep Quality")
plt.xlabel("Phone Usage Before Sleep (minutes)")
plt.ylabel("Sleep Quality Score")

plt.show()
```

```python
sns.pairplot(
    df[['daily_screen_time_hours',
        'sleep_duration_hours',
        'stress_level',
        'physical_activity_minutes',
        'sleep_quality_score']]
)

plt.show()
```

```python

sns.regplot(
    x='daily_screen_time_hours',
    y='stress_level',
    data=df,
    scatter_kws={'alpha':0.4},
    line_kws={'color':'red'}
)

plt.title("Relationship Between Screen Time and Stress Level")
plt.xlabel("Daily Screen Time (Hours)")
plt.ylabel("Stress Level")

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    df['sleep_duration_hours'],
    df['daily_screen_time_hours'],
    df['stress_level'],
    alpha=0.5
)

ax.set_xlabel("Sleep Duration (hours)")
ax.set_ylabel("Screen Time (hours)")
ax.set_zlabel("Stress Level")

ax.set_title("3D Relationship: Sleep, Screen Time, and Stress")

plt.show()
```

```python
df['lifestyle_score'] = (
    df['sleep_duration_hours'] * 2 +
    df['physical_activity_minutes'] * 0.05
) - (
    df['daily_screen_time_hours'] * 1.5 +
    df['caffeine_intake_cups'] * 2
)
plt.figure(figsize=(6,4))

sns.scatterplot(
    x='lifestyle_score',
    y='stress_level',
    data=df,
    alpha=0.5
)

sns.regplot(
    x='lifestyle_score',
    y='stress_level',
    data=df,
    scatter=False,
    color='red'
)

plt.title("Lifestyle Score vs Stress Level")
plt.xlabel("Lifestyle Score")
plt.ylabel("Stress Level")

plt.show()
```

```python
sns.histplot(df['lifestyle_score'], bins=40, kde=True)

plt.title("Distribution of Lifestyle Score")
plt.xlabel("Lifestyle Score")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.scatterplot(
    data=df,
    x="daily_screen_time_hours",
    y="sleep_duration_hours",
    size="stress_level",
    hue="physical_activity_minutes",
    sizes=(20,300),
    palette="viridis",
    alpha=0.7
)

plt.title("Screen Time vs Sleep Duration (Bubble Size = Stress, Color = Physical Activity)")
plt.xlabel("Daily Screen Time (Hours)")
plt.ylabel("Sleep Duration (Hours)")

plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.regplot(
    x="daily_screen_time_hours",
    y="stress_level",
    data=df,
    scatter_kws={'alpha':0.4},
    line_kws={'color':'black'}
)

sns.scatterplot(
    data=df,
    x="daily_screen_time_hours",
    y="stress_level",
    hue="sleep_quality_score",
    palette="coolwarm"
)

plt.title("Lifestyle Risk Map: Screen Time vs Stress")
plt.xlabel("Daily Screen Time (Hours)")
plt.ylabel("Stress Level")

plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

features = [
    'sleep_duration_hours',
    'daily_screen_time_hours',
    'stress_level',
    'physical_activity_minutes',
    'caffeine_intake_cups'
]

scaler = MinMaxScaler()

scaled_values = scaler.fit_transform(df[features])

scaled_df = pd.DataFrame(scaled_values, columns=features)
lifestyle_avg = scaled_df.mean()
import numpy as np

labels = lifestyle_avg.index
values = lifestyle_avg.values

angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)

values = np.concatenate((values,[values[0]]))
angles = np.concatenate((angles,[angles[0]]))

fig = plt.figure(figsize=(4,4))
ax = plt.subplot(111, polar=True)

ax.plot(angles, values, linewidth=2)
ax.fill(angles, values, alpha=0.25)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

plt.title("Average Lifestyle Profile of Users", size=14)

plt.show()
```
